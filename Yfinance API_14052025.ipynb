```python
import yfinance as yf

# List of ticker symbols including AAPL and 7 others
tickers_list = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NFLX", "NVDA", "FB"]

# Create a dictionary to store ticker objects for each company
ticker_data = {ticker: yf.Ticker(ticker) for ticker in tickers_list}

# Iterate over the tickers and fetch data
for ticker, data in ticker_data.items():
    print(f"--- Data for {ticker} ---")
    # General information
    print("Info:")
    print(data.info)
    
    # Earnings calendar (if available)
    print("Calendar:")
    print(data.calendar)
    
    # Analyst price targets (if available)
    print("Analyst Price Targets:")
    print(data.analyst_price_targets)
    
    # Quarterly income statement
    print("Quarterly Income Statement:")
    print(data.quarterly_income_stmt)
    
    # History for the last 1 month
    print("1 Month History:")
    print(data.history(period='1mo'))
    
    # Option chain calls (if options data is available)
    if data.options:
        print("Option Chain (Calls):")
        print(data.option_chain(data.options[0]).calls)
    else:
        print("Options data not available for this ticker.")
    
    print("\n")
```

```python
import yfinance as yf
import pandas as pd

# Define the list of tickers (using "META" instead of "FB")
tickers_list = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NFLX", "NVDA", "META"]

# Prepare a dictionary to store the hourly data for each ticker
hourly_data_dict = {}

for tick in tickers_list:
    print(f"\nProcessing {tick}...")
    
    # Fetch full available historical daily data for the ticker.
    ticker_obj = yf.Ticker(tick)
    data = ticker_obj.history(period='max')
    
    # Ensure the index is in datetime format.
    data.index = pd.to_datetime(data.index)
    
    # Create an hourly date range spanning the full period.
    full_hourly_range = pd.date_range(start=data.index.min(),
                                      end=data.index.max() + pd.Timedelta(hours=23),
                                      freq='H')
    
    # Reindex the daily data to include an entry for every hour (including weekends)
    data_hourly = data.reindex(full_hourly_range)
    
    # Forward fill missing entries, so that missing hours (like during weekends)
    # are filled with the most recent trading day's data (i.e. Friday's data if no weekend data exists)
    data_hourly = data_hourly.fillna(method='ffill')
    
    # Optional: Rename the index to 'Date' for clarity.
    data_hourly.index.name = 'Date'
    
    # Add a new column for the company ticker.
    data_hourly['Company'] = tick
    
    # Store the hourly DataFrame into the dictionary.
    hourly_data_dict[tick] = data_hourly
    
    # Display the last 48 rows to see the most recent two days of hourly data
    print(f"Last 48 rows for {tick}:")
    print(data_hourly.tail(48))
```

```python
import yfinance as yf
import pandas as pd
import os

# Define the list of tickers (using "META" instead of "FB")
tickers_list = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NFLX", "NVDA", "META"]

# Prepare a dictionary to store the hourly data for each ticker
hourly_data_dict = {}

for tick in tickers_list:
    print(f"\nProcessing {tick}...")
    
    # Fetch full available historical daily data for the ticker.
    ticker_obj = yf.Ticker(tick)
    data = ticker_obj.history(period='max')
    
    # Ensure the index is in datetime format.
    data.index = pd.to_datetime(data.index)
    
    # Create an hourly date range spanning the full period.
    full_hourly_range = pd.date_range(start=data.index.min(),
                                      end=data.index.max() + pd.Timedelta(hours=23),
                                      freq='H')
    
    # Reindex the daily data to include an entry for every hour (including weekends)
    data_hourly = data.reindex(full_hourly_range)
    
    # Forward fill missing entries, so that missing hours (like during weekends)
    # are filled with the most recent trading day's data (i.e. Friday's data if no weekend data exists)
    data_hourly = data_hourly.fillna(method='ffill')
    
    # Rename the index to 'Date' for clarity.
    data_hourly.index.name = 'Date'
    
    # Add a new column for the company ticker.
    data_hourly['Company'] = tick
    
    # Store the hourly DataFrame into the dictionary.
    hourly_data_dict[tick] = data_hourly
    
    # Display the last 48 rows to see the most recent two days of hourly data.
    print(f"Last 48 rows for {tick}:")
    print(data_hourly.tail(48))

# Option 2: Combine data from all tickers and save to CSV.
combined_data = pd.concat(hourly_data_dict.values())

# Optional: Reset the index if you want the Date column to become a regular column.
combined_data.reset_index(inplace=True)
combined_data.rename(columns={'index': 'Date'}, inplace=True)

# Save the combined DataFrame to a CSV file
csv_filename = '8_companies_all_data.csv.csv'
combined_data.to_csv(csv_filename, index=False)
print(f"\nCSV file '{csv_filename}' saved successfully.")

# ------------------------------
# Add a code that checks if the file exists and prints a preview.
if os.path.exists(csv_filename):
    print(f"Verified: '{csv_filename}' exists.")
    # Read the CSV file back into a DataFrame.
    df_loaded = pd.read_csv(csv_filename)
    print("\nHere is a preview of the CSV file:")
    print(df_loaded.head(10))
else:
    print(f"Error: '{csv_filename}' not found.")
```
